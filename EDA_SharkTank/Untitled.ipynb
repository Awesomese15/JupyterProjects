```python
#Q10. How many sharks participated in this show and What were their names?

count_of_sharks = [col.replace('_present', '').capitalize() for col in shark_tank_data.columns if col.endswith('_present')]
count_of_sharks
```

```python
#Q11. Find appearance of each sharks? 
def ifPresent(ep,shark):
    if(len(shark_tank_data[(shark_tank_data['episode_number'] == ep) & (shark_tank_data[shark] ==1)])>0):
        return True;
    else:
        return False;
def calculateAppearance(shark):
    count=0
    for ep in shark_tank_data['episode_number'].unique():
        
        if(ifPresent(ep, shark)):
            count += 1
    return count

index =range(6)
data=[{'Ashneer' : calculateAppearance('ashneer_present'),
     'Anupam' : calculateAppearance('anupam_present'),
     'Aman' : calculateAppearance('aman_present'),
     'Namita' :  calculateAppearance('namita_present'),
     'Vineeta' : calculateAppearance('vineeta_present'),
     'Peeyush' : calculateAppearance('peyush_present'),
     'Ghazal' : calculateAppearance('ghazal_present')
     }]
appearance_each_shark = pd.DataFrame(data)
appearance_each_shark.index = ['Total']
appearance_each_shark
```

```python
#Q12. How many entrepreneurs were present? 
entrepreurs = len(shark_tank_data['brand_name'].unique())
entrepreurs
```

```python
#Q13. How many times each shark invested the deal? 
deal_cols = [col for col in shark_tank_data.columns if col.endswith('_deal')]
total_deal = []
for col in deal_cols:
    total_deal.append(shark_tank_data[col].sum())

shark_deal_df = pd.DataFrame({'name': deal_cols, 'Total Deal': total_deal })
shark_deal_df
```

```python
#Q14. Find the equity percent that each sharks gets?
def calCulatePer(value):
    
    try:
        percent = round(100/value,2)
        return percent
    except:
        return 0
 
dl_cols = [col for col in shark_tank_data.columns if col.endswith('_deal')]
dl_cols
shrk_get = shark_tank_data[dl_cols]
shrk_get['euity_percentage'] = shark_tank_data['total_sharks_invested'].apply(calCulatePer)
#shrk_get['euity_percentage'] = shark_tank_data.loc[:['total_sharks_invested']].apply(calCulatePer)
shrk_get
```
