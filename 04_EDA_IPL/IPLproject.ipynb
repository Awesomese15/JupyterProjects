```python
import pandas as pd
```

```python
ipl_data = pd.read_csv('IPL_Dataset.csv')
ipl_data
```

```python
#Q1. What is the maximum number of matches played by an individual player in a season?
max_matches_in_a_season = ipl_data[ipl_data['Mat']==ipl_data['Mat'].max()]
max_matches_in_a_season
```

```python
#Q2. What is the highest average of a player in a season?
avg_highest_in_a_seasion = ipl_data[ipl_data['Avg']==ipl_data['Avg'].max()]
avg_highest_in_a_seasion
```

```python
#Q3. What is the average number of matches played by the players in a season?
avg_match_played_by_players = ipl_data['Mat'].mean().round(2)
avg_match_played_by_players
```

```python
ipl_data = ipl_data.drop(columns = 'POS')
```

```python
#Q4. Please check the correlation between all the features.
corr_matrix = ipl_data.corr(numeric_only=True)
corr_matrix
```

```python
#Q5. Check the list of players who has an average greater than 50 as well strike rate above 120.
pl_avg_n_sr = ipl_data[(ipl_data['Avg']>50) & (ipl_data['SR']>120)]
pl_avg_n_sr[['Player', 'Avg', 'SR']]
```

```python
#Q6. Please check the list of players who has an average greater than 40 and balls faced above 100.
pl_abv40_n_bf_abv100 = ipl_data[(ipl_data['Avg']>40) & (ipl_data['BF']>100)]
pl_abv40_n_bf_abv100[['Player', 'Avg', 'BF']]
```

```python
#Q7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.
pl_run300_avg45_sr130 = ipl_data[(ipl_data['Runs']>300) & (ipl_data['Avg']>45) &(ipl_data['SR']>130)]
pl_run300_avg45_sr130
```

```python
#Q8. Players who scroed alteast one century in this season.
pl_atleast_one_100 = ipl_data[ipl_data['100']>=1]
pl_atleast_one_100
```

```python
#Q9. Players who scored atleast 4 half centuries in this season.
pl_atleast_4_half_century = ipl_data[ipl_data['50']>=4]
pl_atleast_4_half_century
```

```python
#Q10. Check the list of players who hit more than 45 boundARIES and more than 10 sixes in this season.
ipl_data[(ipl_data['4s'] >45)&(ipl_data['6s'] >10)]
```

```python
#Q11. Plot a histogram of number of matches played in a season by players.
import matplotlib.pyplot as plt
```

```python
ipl_data.hist('Mat', bins=10, edgecolor='black')
plt.xlabel('Matches')
plt.ylabel('Frequency')
plt.title('Histogram of Matches played by a player in a season')
plt.show()
```

```python
#Plot the histogram of balls faced by players.
ipl_data.hist(column='BF', bins=20, edgecolor='black')
plt.xlabel('Balls faced')
plt.ylabel('Frequency')
plt.title('Histogram for balls faced by a player')
```

```python
#Q13. Top 20 players with most runs in a season.
top_20_batter = ipl_data.sort_values(by='Runs', ascending=False)
top_20_batter.head(20)
```

```python
#Q14. Bar chart for top 20 players with best strike rate in a season.
top_20_sr = ipl_data.nlargest(20 ,columns= 'SR')
#top_20_sr = ipl_data.sort_values(by='SR', ascending= False).head(20)
plt.figure(figsize=(10, 6))
plt.bar(top_20_sr['Player'], top_20_sr['SR'],color='blue')
plt.xticks(rotation=90)
plt.xlabel('Player')
plt.ylabel('Strik rate')
plt.show()
```

```python
#Q15. Top 30 players with highest number of sixes.
top_30_six = ipl_data.nlargest(30, columns='6s')
top_30_six
```

```python
#Q16. Scatter plot of runs scored by a player v/s balls faced in a season.
plt.scatter(ipl_data['Runs'], ipl_data['BF'],color='blue', marker='o', label='Data Points')
plt.xlabel('Runs')
plt.ylabel('Balls faced')
plt.title('Runs vs Balls Faced')

plt.legend()
plt.grid(True)

plt.show()
```

```python
#Q17. KDE plot for averages on a season.
import seaborn as sns
```

```python
sns.kdeplot(ipl_data['Avg'],color='blue', label='KDE Plot for avgerage')
plt.xlabel('Avg')
plt.ylabel('Density')
plt.title('KDE Plot for avgerage')

plt.legend()  # Show legend if you added a label

plt.show()
```
